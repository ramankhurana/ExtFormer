```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Function to extract the first metric (MSE) from a .npy file
def extract_mse(file_path):
    data = np.load(file_path)
    return data[1]  # Assuming the MSE is the first metric

# Path to the results directory
results_dir = '../results-k8/'

# List all experiment directories
experiment_dirs = [d for d in os.listdir(results_dir) if os.path.isdir(os.path.join(results_dir, d))]

# Initialize a list to store the data
data = []
dataset = "Divvy"
dataset2 = dataset+"_"+dataset
# Iterate over each experiment directory
for exp_dir in experiment_dirs:
    parts = exp_dir.split('_')
    model_name = parts[6]  # Extracting the model name
    seq_len = parts[4]  # Extracting the sequence length
    pred_len = parts[5]  # Extracting the prediction length
    experiment_name = model_name+"_SL_"+seq_len+"_PL_"+pred_len+"_"+dataset
    print (experiment_name)
    none_metric_path = os.path.join(results_dir, exp_dir, f"{model_name}_{dataset2}_{seq_len}_{pred_len}_NONE_metrics.npy")
    static4_metric_path = os.path.join(results_dir, exp_dir, f"{model_name}_{dataset2}_{seq_len}_{pred_len}_static4_metrics.npy")

    # Check if both files exist
    if os.path.exists(none_metric_path) and os.path.exists(static4_metric_path):
        mse_none = extract_mse(none_metric_path)
        mse_static4 = extract_mse(static4_metric_path)

        # Add the data to the list
        data.append({
            'Experiment': experiment_name,
            'MSE_None': mse_none,
            'MSE_Static4': mse_static4
        })

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming 'df' is your original DataFrame and it has columns for each experiment
# and 'mse_none' and 'mse_static4' are the columns with the MSE values.

# Calculate the relative change
df['relative_change'] = (df['MSE_Static4'] - df['MSE_None']) / df['MSE_None']
df['relative_change_capped'] = df['relative_change'].clip(-0.1, 0.1)

column_name = 'relative_change_capped'
num_positive_changes = (df[column_name] > 0).sum()
num_negative_changes = (df[column_name] < 0).sum()


# Now, create a new DataFrame for plotting
df_plot = df[['Experiment', column_name]].copy()
df_plot['Experiment'] = df_plot['Experiment'].astype(str)  # Ensure that Experiment is of type string

# Plotting the relative change
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Experiment', y=column_name, data=df_plot)
plt.xticks(rotation=90)
plt.ylabel('Relative Change (%)')
plt.title('Relative MSE Change in Different Experiments')
plt.axhline(0, color='black', linewidth=0.8)  # Add a line at 0 for reference

plt.text(x=0, y=df['relative_change'].max()*0.95, s=f'Positives: {num_positive_changes}', color='green', fontsize=12)
plt.text(x=0, y=df['relative_change'].max()*0.8, s=f'Negatives: {num_negative_changes}', color='red', fontsize=12)

# Formatting y-axis as percentage
from matplotlib.ticker import PercentFormatter
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configurable parameters
dataset_name = 'M5'  # Change this to switch datasets (e.g., 'M5')

# Filter the DataFrame for the specified dataset
df_filtered = df[df['dataset'] == dataset_name]

# Calculate the percentage change w.r.t. NONE for each static case
percentage_changes = []
for (model, seq_len, pred_len), group in df_filtered.groupby(['model', 'seq_len', 'pred_len']):
    none_mse = group[group['static'] == 'NONE']['MSE'].values[0]
    for _, row in group.iterrows():
        if row['static'] != 'NONE':
            percent_change = ((row['MSE'] - none_mse) / none_mse) * 100
            percentage_changes.append({
                'model': model,
                'seq_len': seq_len,
                'pred_len': pred_len,
                'static': row['static'],
                'Percentage Change': percent_change
            })

# Convert the list of dictionaries to a DataFrame
df_percentage_changes = pd.DataFrame(percentage_changes)


# Assuming df_percentage_changes is already created as per previous instructions

# Sort the DataFrame to ensure the static cases are ordered consistently across models
df_percentage_changes.sort_values(by=['model', 'static'], inplace=True)

# Use catplot to create a plot with separate sections for each model
g = sns.catplot(
    data=df_percentage_changes, 
    kind="bar", 
    x="model", 
    y="Percentage Change", 
    hue="static", 
    palette="viridis",
    height=6, 
    aspect=2,
    legend_out=True
)

# Improve plot aesthetics and layout
g.despine(left=True)
g.set_axis_labels("", "Percentage Change")
g.legend.set_title("Static Case")
g.set_titles("{col_name} Dataset")
plt.axhline(0, color='gray', linestyle='--')  # Reference line at 0%

# Adjust legend position
#plt.legend(title='Static Case', bbox_to_anchor=(1.05, 1), loc='upper left')

# Final adjustments
plt.title(f'Percentage Change w.r.t. NONE across Models for {dataset_name} Dataset')
plt.tight_layout()

plt.show()
```
