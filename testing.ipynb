```python
import numpy as np
import matplotlib.pyplot as plt

# Create a time axis
t = np.linspace(0, 12, 100)

# Calculate the Fourier series coefficients
A0 = 15.8
A1 = 13.0
A2 = 3.2
A3 = 1.0
A4 = 0.3

# Calculate the Fourier series
T = A0 + A1 * np.cos(2 * np.pi * t / 12) + A2 * np.cos(4 * np.pi * t / 12) + A3 * np.cos(6 * np.pi * t / 12) + A4 * np.cos(8 * np.pi * t / 12)

# Plot the time series
plt.plot(t, T)

# Set the axis labels and title
plt.xlabel("Month")
plt.ylabel("Average Monthly Temperature (Â°C)")
plt.title("Average Monthly Temperature in the United States")

# Show the plot
plt.show()
```
